# **4 Data visualization**

## Pie chart of the percentage of the os the laptop is using from all laptops

```python
def label_function(val):
    return f'{val / 100 * len(df):.0f}\n{val:.0f}%'

fig, ax1 = plt.subplots(ncols=1, figsize=(30, 30))

df.groupby('OS').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 20},
                                  colors=['tomato', 'gold', 'skyblue'], ax=ax1)

ax1.set_ylabel('Per OS', size=22)
plt.legend()
plt.tight_layout()
plt.show()
```

## Bar chart of the average weights of each type laptop 

```python
x=df['Type']
y=df['Weight']
plt.subplots(figsize=(18,5))
plt.bar(x,y,color='skyblue')
plt.show()
```

## Scattered plot that represent the price of the laptop depending on its type and company

```python
plt.subplots(figsize=(18,15))
sns.scatterplot(x='Company', y='Price', hue='Type', data=df) 
plt.show()
```
